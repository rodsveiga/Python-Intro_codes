## Visual exploratory data analysis

### Weekly average temperature and visibility

Is there a correlation between temperature and visibility? Now we are going to plot the weekly average temperature and visibility as subplots. To do this, we need to first select the appropriate columns and then resample by week, aggregating the mean.

In addition to creating the subplots, we will compute the Pearson correlation coefficient using `.corr()`. The Pearson correlation coefficient, known also as Pearson's r, ranges from -1 (indicating total negative linear correlation) to 1 (indicating total positive linear correlation). A value close to 1 here would indicate that there is a strong correlation between temperature and visibility.

```python
# Select the `visibility` and `dry_bulb_faren` columns and resample them.

weekly_mean = df[['visibility', 'dry_bulb_faren']].resample('W').mean()

# When we slice on more than one features, they have to be passed in the 
# argument as a list.
```

```python
weekly_mean.head()
```

```python
# The Pearson correlation coefficient.

weekly_mean.corr()
```

```python
# Plotting `weekly_mean` with `subplots=True`.

weekly_mean.plot(subplots=True)
plt.show()
```

### Daily hours of clear sky

In a previous task, we analyzed the `sky_condition` column to explore the difference in temperature on sunny days compared to overcast days. Recall that a `sky_condition` of `CLR` represents a sunny day. Now we are goint to explore sunny days in greater detail. Specifically, using a box plot to visualize the fraction of days that are sunny.

The `sky_condition` column is recorded many times in a day. We will resample this column appropriately such that it can be extracted the number of sunny hours in a day and the number of total hours. Then, we can divide the number of sunny hours by the number of total hours, and generate a box plot of the resulting fraction.

```python
# In order to create the fraction, we are going to use sum of booleans. Below
# we create a Boolean series for sunny days

sunny = df['sky_condition'] == 'CLR'
sunny.head()
```

```python
# Then we resample the Boolean Series by day and compute the `sum`. As `True` counts
# as 1 and `False` as 0, we are just count how many sunny measures daily.

sunny_hours = sunny.resample('D').sum()
sunny_hours.head()
```

```python
# Then we resample the Boolean Series by day and compute the `sum`. Now 
# we are just count counting the total measured of `sky_condition` daily.

total_hours = sunny.resample('D').count()
total_hours.head()
```

```python
# To get the fraction required we just divided the sunny measurements by the total.

sunny_fraction = sunny_hours / total_hours
```

```python
# A box plot of the sunny fraction.

sunny_fraction.plot(kind='box')
plt.show()
```

### Heat or humidity

Dew point is a measure of relative humidity based on pressure and temperature. A dew point above 65 is considered uncomfortable while a temperature above 90 is also considered uncomfortable.

We will now explore the maximum temperature and dew point of each month. The columns of interest are `dry_bulb_faren` and `dew_point_faren`. After resampling them appropriately to get the maximum temperature and dew point in each month, we will generate a histogram of these values as subplots. Uncomfortably, we are gointo to notice that the maximum dew point is above 65 every month! 

```python
monthly_max = df[['dew_point_faren', 'dry_bulb_faren']].resample('M').max()
monthly_max.head()
```

```python
monthly_max.plot(kind='hist', bins=8, alpha=0.5, subplots=True)
plt.show()
```

### Probability of high temperatures

The year 2011 was hotter than the climate normals for the previous thirty years. In thi final task, we will compare the the maximum temperature in August 2011, which is given by `august_max=95.3` against that of the August 2010 climate normals. More specifically, we are going to use a CDF plot to determine the probability of the 2011 daily maximum temperature in August being above the 2010 climate normal value. To do this, we will leverage the data manipulation, filtering, resampling, and visualization skills we have acquired throughout this code.

Once you've generated the CDF, notice how it shows that there was a 50% probability of the 2011 daily maximum temperature in August being 5 degrees above the 2010 climate normal value!

```python
august_max = 95.3

# Resampling the August 2011 temperatures in `df` and aggregating
# the maximum value.

august_2011 = df.loc['2011-Aug', 'dry_bulb_faren'].resample('D').max()
august_2011.head()
```

```python
# Filtering out days in `august_2011` where the value exceeded august_max.

august_2011_high = august_2011[august_2011 > august_max]
```

```python
# CDF plot.

august_2011_high.plot(kind='hist', normed=True, cumulative=True, bins=25)
plt.show()
```

Note the CDF shows that there was a 50% probability of the 2011 daily maximum temperature in August being 5 degrees above the maximum teperature `august_max`.
